```python
#bai toan phan loai chu viet tay - ung voi cac so thap phan
%matplotlib inline
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
digits = load_digits()
#-------------------------------
dir(digits)
#-------------------------------
print(digits.data[0])
```

```python
plt.gray() #thu in data tat ca 10 so: 0-9
for i in range(10):
    plt.matshow(digits.images[i])
```

```python
print(digits.target[0:10]) #target phan loai
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.2)
lo = LogisticRegression()
lo.fit(x_train, y_train)
lo.score(x_test, y_test) #danh gia mo hinh

```

```python
#plt.matshow(digits.images[67])
print(digits.target[67])
pre = lo.predict(x_test)
```

```python
#danh gia mo hinh bang confusion matrix
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test, pre) #xuat o dang matrix
#-> WTF is this? right ? let's see in a graph, easier!
```

```python
#bieu do de hieu confusion matrix
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cf, annot = True)
plt.xlabel('Pre')
plt.ylabel('Truth')
```
