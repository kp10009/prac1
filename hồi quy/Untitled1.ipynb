```python
import time 
import numpy as np
n_val = []
time_end = []
```

```python
class Board:
    def __init__(self, n) :
        self.N = n

    def Draw_board(self, position : np.array):
        for i in range(self.N):
            for j in range(self.N):
                if j  == position[i]:
                    print('|', end= 'Q')
                else:
                    print('|', end= ' ')
            print('|')
class DFS:
    def __init__(self, n):
        self.n = n
        self.solutions = []

    def is_safe(self, board, row, col):
        # Kiểm tra xem vị trí đặt quân hậu có hợp lệ hay không
        for i in range(row):
            if board[i] == col or \
               board[i] - i == col - row or \
               board[i] + i == col + row:
                return False
        return True

    def solve_nqueens(self):
        def dfs(row, board):
            if len(self.solutions) == 1:
                return
            if row == self.n:
                self.solutions.append(np.array(board))
                return
            for col in range(self.n):
                if self.is_safe(board, row, col):
                    board.append(col)
                    dfs(row + 1, board)
                    board.pop()

        dfs(0, [])
        return self.solutions
import time 
n = int(input("Nhập số quân hậu: "))
# store starting time 
begin = time.time() 
 
# program body starts 
Board = Board(n)
solution = DFS(n)
solution.solve_nqueens()
print("Số lời giải:",len(solution.solutions))
for i, position in enumerate(solution.solutions):
    print("Solution", i + 1, ':')
    Board.Draw_board(position)
# program body ends 
 
time.sleep(1) 
# store end time 
end = time.time() 
a = round(end-begin, 5)
print(a)
import pandas as pd
n_val = np.append(n_val, n)

time_end = np.append( time_end, a)
# print(n_val.shape)
# print(time_end.shape)
df = pd.DataFrame({'n' : n_val, 'time' : time_end})
df
```
